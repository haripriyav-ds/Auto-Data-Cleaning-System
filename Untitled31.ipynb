```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# STEP 1: Load Dataset
# -------------------------------
df = pd.read_csv(r"C:\Users\ELCOT\Downloads\titanic\train.csv")

print("🔹 Original Dataset Preview:")
print(df.head())

# -------------------------------
# STEP 2: Basic Info
# -------------------------------
print("\n🔹 Dataset Info:")
print(df.info())

print("\n🔹 Missing Values Before Cleaning:")
print(df.isnull().sum())

# -------------------------------
# STEP 3: Handle Missing Values
# -------------------------------
for col in df.columns:
    if df[col].dtype == 'object':
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        df[col].fillna(df[col].mean(), inplace=True)

print("\n✅ Missing Values After Cleaning:")
print(df.isnull().sum())

# -------------------------------
# STEP 4: Visualization (Before Cleaning Outliers)
# -------------------------------
numeric_cols = df.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    plt.figure()
    df[col].hist()
    plt.title(f"Before Cleaning - {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

# -------------------------------
# STEP 5: Remove Outliers (IQR Method)
# -------------------------------
def remove_outliers(data):
    for col in data.select_dtypes(include=[np.number]).columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        
        data = data[(data[col] >= lower) & (data[col] <= upper)]
    return data

df = remove_outliers(df)

print("\n🔹 Dataset Shape After Outlier Removal:", df.shape)

# -------------------------------
# STEP 6: Visualization (After Cleaning)
# -------------------------------
for col in numeric_cols:
    plt.figure()
    df[col].hist()
    plt.title(f"After Cleaning - {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

# -------------------------------
# STEP 7: Encode Categorical Data
# -------------------------------
df = pd.get_dummies(df, drop_first=True)

print("\n🔹 Dataset After Encoding:")
print(df.head())

# -------------------------------
# STEP 8: Save Cleaned Dataset
# -------------------------------
df.to_csv("cleaned_titanic.csv", index=False)

print("\n✅ Data Cleaning Completed Successfully!")
print("📁 Cleaned file saved as 'cleaned_titanic.csv'")
```
